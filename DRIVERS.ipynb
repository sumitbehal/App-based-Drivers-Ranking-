```python
#the more stops per minute ranks top. the less stops per mile ranks top
import pandas as pd
import numpy as np
df=pd.read_csv('driver (1).csv')
```

This algorithm is designed to find the rankings of drivers considering three features.
1.No. of Stops
2.Time Taken
3.Distance Travelled 

```python
df.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6'],axis=1,inplace=True)
```

```python
df['Stops/Minute']=df['Stops']/df['Total Time']
df['Stops/Mile']=df['Stops']/df['Total Distance']
```

```python
#Scaling the feature
a_max=max(df['Stops/Minute'].values)
a_min=min(df['Stops/Minute'].values)
b_max=max(df['Stops/Mile'].values)
b_min=min(df['Stops/Mile'].values)
```

```python
df['A']=(df['Stops/Minute']-a_min)/(a_max-a_min)
```

```python
df['B']=(b_max-df['Stops/Mile'])/(b_max-b_min)
```

```python
df['C']=df['A']+df['B']
```

```python
df=df.sort_values(['C'],ascending=False)
```

```python
df.index = pd.RangeIndex(start=0, stop=19, step=1)
```
